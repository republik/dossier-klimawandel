```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import csv
%matplotlib inline
plt.rcParams["figure.figsize"] = [15, 4]
```

### Land-Ocean Temperature Index

Globale Temperatur relativ zum langjährigen Durchschnitt 1951–1980

Quelle: [NASA](https://climate.nasa.gov/vital-signs/global-temperature/)

```python
warming = pd.read_csv('./datasets/Land-Ocean-Temperature-Index.txt', skiprows=[0,1,2,4], sep='\s+')
warming.plot(kind='line', x='Year', y='Lowess(5)');
```

```python
warming_display = pd.DataFrame(warming[['Year', 'Lowess(5)']].values, columns=['year', 'value'])
warming_display['year'] = warming_display['year'].astype('int32')
warming_display.to_csv('./warming_global.csv', index=False)
```

### CO₂-Emissionen

#### Menge

Weltweit, Mio. Tonnen pro Jahr

Quelle: [Carbon Atlas](http://www.globalcarbonatlas.org/en/CO2-emissions)

Gesehen bei: 
 * [The Guardian](https://www.theguardian.com/environment/2019/may/07/us-hotbed-climate-change-denial-international-poll)

```python
emissions = pd.read_csv('./datasets/globalcarbonatlas_export_20190521_1217.csv', sep=';', index_col='Year')
```

```python
emissions_selected = emissions[['China', 'United States of America', 'EU28']]
emissions_selected.loc[:,'Andere'] = emissions['World'] - emissions_selected.sum(axis=1)
emissions_selected = emissions_selected.rename(columns={'United States of America': 'USA'})
emissions_selected.to_csv('./emissions_selected.csv')
```

```python
emissions_selected.plot(kind='bar', stacked=True);
```

```python
emissions_rows = []
for index, row in emissions_selected.iterrows():
    for item in row.iteritems():
        key, val = item
        emissions_rows.append([index,key,val])

emissions_display = pd.DataFrame(emissions_rows, columns=['year','label','value'])
emissions_display.to_csv('./emissions_global.csv', float_format='%.4f', index=False)
```
